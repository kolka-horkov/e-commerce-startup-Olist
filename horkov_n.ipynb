```python
def fun(row):
    if row['diff'].days <= 30:
        return 0
    if 30 <row['diff'].days <= 60:
        return 1
    if 60 <row['diff'].days <= 90:
        return 2
    if 90 <row['diff'].days <= 120:
        return 3
    if 120 <row['diff'].days <= 150:
        return 4
    if 180 <row['diff'].days <= 210:
        return 5
    if 210 <row['diff'].days <= 240:
        return 6
    if 240 <row['diff'].days <= 270:
        return 7
    if 270 <row['diff'].days <= 300:
        return 8
    if 300 <row['diff'].days <= 330:
        return 9
    if 330 <row['diff'].days <= 360:
        return 10
    if 360 <row['diff'].days <= 390:
        return 11
    if 3900 <row['diff'].days <= 420:
        return 12
```

# Используя python, построй RFM-сегментацию пользователей, чтобы качественно оценить свою аудиторию.
В кластеризации можешь выбрать следующие метрики: 
    R - время от последней покупки пользователя до текущей даты, 
    F - суммарное количество покупок у пользователя за всё время, 
    M - сумма покупок за всё время. Подробно опиши, как ты создавал кластеры. 
Для каждого RFM-сегмента построй границы метрик recency, frequency и monetary для интерпретации этих кластеров. 
Пример такого описания: RFM-сегмент 132 (recency=1, frequency=3, monetary=2)
имеет границы метрик recency от 130 до 500 дней, 
frequency от 2 до 5 заказов в неделю, monetary от 1780 до 3560 рублей в неделю.

```python
def mon(row):
    if row['price'] <= 25 :
        return 1
    if 25 <row['price'] <= 150 :
        return 2
    if row['price'] > 150 :
        return 3
# описание функции, которые будет делить наш датафрейм на категории в зависимости от суммы покупок
```

```python
df_6['M']= df_6.apply(mon, axis=1) # применение функции и запись в колонку значении категори
```

Границы метрик:
    monetary: от 0 до 25 рублей за весь период = 1
              от 25 до 150 рублей за весь период =2 
              от 150 и выше =3
Группа 1 - это 7000 человек ( 8% от всего количества), которые совершали по 1 незначильной покупки.
Группа 2 - это все остальные люди.
Группа 3 - это 21000 человек (22% от всего количества), которые в основном покупают 1 покупке, которая находятся 
в дорогом сегменте и только 5% покупают больше вещи.

```python
def quan(row):
    if row['quantity'] == 1 :
        return 1
    if row['quantity'] == 2 :
        return 2
    if row['quantity'] > 2 :
        return 3
# описание функции, которые будет делить наш датафрейм на категории в зависимости от количество покупок за все время
```

```python
df_6['F']= df_6.apply(quan, axis=1) # применение функции и запись в колонку значении категори
```

Границы метрик:
    frequency: 1 покупка за весь период = 1
               2 покупки за весь период = 2 
               больше 2 покупок ха весь период = 3
Группа 1 - это группа людей, которая один раз приобрела товар и больше не возращалась.
Группа 2 - это группа людей, которой видимо понравилось наше приложение и эту группу нужно ппробовать удержать
Группа 3 - это группа людей, которая распробовала наше приложение и во всю ей пользуется.
